```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
sent=['this is the data science class with SL',
       'AI is the next class',
       'and this is the next class',
       'and this is fourth class']
```

#### Vocabulary Size

```python
vocab_size=10000
```

#### One hot representation

```python
onehot_repr=[one_hot(words,vocab_size)for words in sent]
print(onehot_repr)
```

```python
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences # To ensure that all sequences have the same length, padding is applied.
from tensorflow.keras.models import Sequential
```

```python
import numpy as np
```

```python
embeded_docs=pad_sequences(onehot_repr,padding='pre')
embeded_docs
```

```python
sent_lenght=10
embeded_docs=pad_sequences(onehot_repr,padding='pre',maxlen=sent_lenght)# The padding='post' argument indicates that padding should be added at the end of the sequences.
embeded_docs
```

```python
dim=15 #features
```

```python
model=Sequential()
model.add(Embedding(vocab_size,dim,input_length=sent_lenght))
model.compile('adam','mse')
```

```python
model.summary()
```

```python
model.predict(embeded_docs)
```
