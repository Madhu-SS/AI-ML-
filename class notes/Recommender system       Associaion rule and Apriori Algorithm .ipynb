# Association rule and Apriori algorithm

```python
from mlxtend.frequent_patterns import association_rules,apriori
```

```python
tit=pd.read_csv('Titanic.csv')
```

```python
tit
# objective was to find the survival or not
```

```python
tit1=pd.get_dummies(tit)
```

```python
tit1
```

```python
freq_pattern=apriori(tit1,min_support=0.5,use_colnames=True)
```

```python
freq_pattern
```

```python
 tit.Age.value_counts()
```

```python
2092/2201
```

```python
association_rules(freq_pattern,metric='confidence',min_threshold=0.7)
```
