## crosstab

```python
pd.crosstab(index=car['Type'],columns=[car['Weight'],car['EngineSize']])
```

## groupby

```python
car.groupby(car['Origin']).get_group('USA')
```

```python
car['Origin'].unique()
```

```python
car.groupby([car['Origin'],car['name_of_car']])[['Weight','EngineSize']].min()
```

```python
car.groupby([car['Origin'],car['name_of_car']])[['Weight','EngineSize']].max()
```

```python
car.groupby([car['Origin'],car['name_of_car']])[['Weight','EngineSize','Horsepower']].agg([max,min,np.mean])
```

```python
car[['Weight','EngineSize','Horsepower']].groupby([car['Origin'],car['name_of_car']]).agg([max,min,np.mean])
```
