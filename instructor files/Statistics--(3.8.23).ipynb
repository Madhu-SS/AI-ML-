```python
import scipy.stats as stats
import pandas as pd
```

```python
stat1=stats.uniform.rvs(size=1000,loc=100,scale=80)
stat1
```

```python
stat_data=pd.DataFrame(stat1)
stat_data
```

```python
stat_data.plot(kind='density',figsize=(6,8))
```

```python
stat2=stats.norm.rvs(size=1000,loc=100,scale=80)
stat2
```

```python
stat2_data= pd.DataFrame(stat2)
```

```python
stat2_data.plot(kind='density')
```

1.	The boys in the class are having mean weight= 9.5kg
Standard deviation=1.1kg
Calculate the percentage(%) of the boys in the following:
1.	Weight less than 8.4kg
2.	Between 7.3kg and 11.7kg
3.	More than 12.8


```python
stats.norm.cdf(8.4,loc=9.5,scale=1.1)
```

```python
p1= stats.norm.cdf(7.3,loc=9.5,scale=1.1)
p1
```

```python
p2= stats.norm.cdf(11.7,loc=9.5,scale=1.1)
p2
```

```python
p2-p1
```

```python
p3= stats.norm.cdf(12.8,loc=9.5,scale=1.1)
p3
```
