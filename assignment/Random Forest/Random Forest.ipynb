### A cloth manufacturing company is interested to know about the segment or attributes causes high sale. 
### Approach - A Random Forest can be built with target variable Sales (we will first convert it in categorical variable) & all other variable will be independent in the analysis.  


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import LabelEncoder
```

```python
df=pd.read_csv('Company_Data.csv')
```

```python
df
```

```python
df.Sales.mean()
```

```python
df['sales']='small'
df.loc[df['Sales']>=7.49,'sales']='large'
df.drop(['Sales'],axis=1)
```

```python
target=df[['sales']]
```

```python
lab_enc=LabelEncoder()
```

```python
target['sales']=lab_enc.fit_transform(target['sales'])
```

```python
features=df.drop(['sales'],axis=1)
```

```python
features
```

```python
features['Education']=lab_enc.fit_transform(features['Education'])
features['ShelveLoc']=lab_enc.fit_transform(features['ShelveLoc'])
```

```python
features=pd.get_dummies(features)
```

```python
features
```

```python
x_train,x_test,y_train,y_test=train_test_split(features,target,train_size=0.75,random_state=80)
```

```python
ran_forest=RandomForestClassifier( n_estimators=100)
```

```python
ran_forest.fit(x_train,y_train)
```

```python
y_pred=ran_forest.predict(x_test)
```

```python
accuracy_score(y_test,y_pred)
```

```python
confusion_matrix(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```

### Use Random Forest to prepare a model on fraud data 
### treating those who have taxable_income <= 30000 as "Risky" and others are "Good"


```python
df=pd.read_csv('Fraud_check.csv')
```

```python
df
```

```python
df['Taxable.Income']=pd.cut(df['Taxable.Income'],bins=[0,30000,100000],labels=["Risky","Good"])
```

```python
df
```

```python
features=df.drop({'Taxable.Income'},axis=1)
```

```python
features
```

```python
features['Marital.Status']=lab_enc.fit_transform(features['Marital.Status'])
```

```python
features=pd.get_dummies(features)
features
```

```python
target=df[['Taxable.Income']]
```

```python
target['Taxable.Income']=lab_enc.fit_transform(target['Taxable.Income'])
```

```python
x_train,x_test,y_train,y_test=train_test_split(features,target,train_size=0.75,random_state=80)
```

```python
rand_forest=RandomForestClassifier(n_estimators=64)
```

```python
rand_forest.fit(x_train,y_train)
```

```python
y_pred=rand_forest.predict(x_test)
```

```python
accuracy_score(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```

```python
confusion_matrix(y_test,y_pred)
```

```python
n=range(1,100)
```

```python
params=dict( n_estimators=n)
```

```python
grid_search=GridSearchCV(rand_forest,params)
```

```python
grid_search.fit(x_train,y_train)
```

```python
grid_search.best_params_
```
